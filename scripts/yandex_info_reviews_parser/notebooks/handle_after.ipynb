```python
true_row = sber_parse[sber_parse['true_row'] == True]
false_row = sber_parse[sber_parse['true_row'] == False]

true_cities = list(np.unique(true_row['city']))
error_cities =  list(np.unique(false_row[~false_row['city'].isin(true_cities)]['city']))
```

```python
def remove_cities(cities, bank_name):
    # удаление
    info_path  =f'info_output/{bank_name}/'
    links_path  =f'links/{bank_name}/'

    info_counter = 0
    links_counter = 0

    for f in cities:
        try:
            os.remove(info_path + f + '_info.csv')
            info_counter += 1
        except Exception as e:
            print(f'error in remove {info_path}{f}_info.csv: {e}')
        try:
            os.remove(f'{links_path} link_{f}.pkl')
            links_counter += 1
        except Exception as e:
            print(f'error in remove {links_path}link_{f}.pkl: {e}')

    print(f'{info_counter} info was removed')
    print(f'{links_counter} links was removed')
```

```python
remove_cities(error_cities, 'alfa_bank')
```
